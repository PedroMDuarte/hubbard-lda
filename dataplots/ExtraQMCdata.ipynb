```python
# The order of the columns is 
# mu, dens, spi, spi_err 

# It needs to be changed to 
# mu, dens, entr, spi/n  

import os

MUCOL = 0
DENSCOL = 1 

datlist = [
    ( np.loadtxt('../QMC_Extra/U7_T0.36_6x6x6_140624.dat'),  7.0, 0.360,  'None' ),
    ( np.loadtxt('../QMC_Extra/U8_T0.40_6x6x6_140626.dat'),  8.0, 0.400,  'None' ),
    ( np.loadtxt('../QMC_Extra/U10_T0.50_6x6x6_140630.dat'), 10.0, 0.500,  'QMC' ),
    ( np.loadtxt('../QMC_Extra/U12_T0.50_6x6x6_140630.dat'), 12.0, 0.500, 'NLCE' ),
  ]

for dat, Uval, Tval, previous in datlist:
    print 
        
    # The SPI data will be reduced in the number of 
    a = np.empty( (dat.shape[0]*2-1, 4))
    
    mu = dat[:,0]
    dens = dat[:,1] 
    spi = dat[:,2] 
    
    a[:,0] = np.concatenate(( mu, (-1.*mu[:-1])[::-1] )) + Uval/2. # chemical potential adjusted for particle-hole
    a[:,1] = np.concatenate(( dens, 2. - dens[:-1][::-1] ))  # density
    a[:,2] = np.zeros_like( a[:,1] )  # entropy 
    a[:,3] = np.concatenate(( spi, spi[:-1][::-1] )) / a[:,1] # spi 
    
    if previous == 'NLCE':
        fname = '../NLCE8_Final/U{:02d}/T{:.2f}.dat'.format(int(Uval),Tval)
        dat_nlce = np.loadtxt(fname)
        
        subset = dat_nlce[:,1] > 0.02  
        dat_nlce = dat_nlce[subset ]

        b = np.empty((dat_nlce.shape[0],4))

        b[:,0] = dat_nlce[:,0]  # chemical potential
        b[:,1] = dat_nlce[:,1]  # density 
        b[:,2] = dat_nlce[:,2]  # entropy 
        spi_over_n = dat_nlce[:,4]
        b[:,3] = spi_over_n
        
        # For NLCE complements just use only QMC data
        c = a 
        #c = np.vstack(( a, b ))
        #c = c[np.argsort(c[:,0])]
        
    elif previous == 'QMC':
        fname = '../QMC_Final/U{:02d}/T{:.3f}.dat'.format(int(Uval),Tval)
        dat_qmc = np.loadtxt(fname)
        
        # The SPI data will be reduced in the number of 
        b = np.empty( (dat_qmc.shape[0], 4))
        b[:,0] = dat_qmc[:,0] # chemical potential
        b[:,1] = dat_qmc[:,1] # density
        b[:,2] = np.zeros_like( dat_qmc[:,0]) # entropy
        b[:,3] = dat_qmc[:,3]
        c = np.vstack(( a, b ))
        c = c[np.argsort(c[:,0])]
        
    else:
        b = None 
        c = a 

    savedir = '../COMB_Final_Spi/U{:02d}/'.format(int(Uval))
    if not os.path.exists(savedir):
        os.makedirs(savedir)
    fnameout = savedir + 'T{:.3f}.dat'.format(Tval)
    
    np.savetxt( fnameout, c, fmt='%.3g')
    #print Uval, Tval, fname 
```
