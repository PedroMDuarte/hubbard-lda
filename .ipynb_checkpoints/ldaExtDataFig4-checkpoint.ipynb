```python
import glob 
from colorChooser import rgb_to_hex, cmapCycle

def fx(x):
    return x * 11.1 /290. 

expdat = np.loadtxt('dataplots/SPI/experiment_data.txt')



from matplotlib import rc
rc('font', **{'family':'serif'})
rc('text', usetex=True)


gs = matplotlib.gridspec.GridSpec(1,1,\
      wspace=0.35, hspace=0.05,\
      left=0.20, right=0.86, bottom=0.18, top=0.93)


figS = plt.figure( figsize(3.2,2.4))
axS = figS.add_subplot( gs[0] )


aSs = [200, 290, 380, 470]


Tsets = []
for aS in aSs:
    fdats  = sorted(glob.glob('dataplots/SPI/{:03d}/*.dat'.format(int(aS))))
    Tspis  = [ float(f.split('Tspi')[1].split('.dat')[0]) for f in fdats ] 
    print aS
    print Tspis
    Tsets.append( set(Tspis))
    
    
    
Tcommon = sorted(set.intersection(*Tsets))
print "Common:"
print Tcommon

for i,Tspi in enumerate(Tcommon):
    
    #color = cmapCycle( matplotlib.cm.bone, float(i+3), lbound=0, ubound=float(len(Tcommon)+4))
    #color = cmapCycle( matplotlib.cm.winter, float(i), lbound=0, ubound=float(len(Tcommon)+0))
    color = cmapCycle( matplotlib.cm.cool, float(i), lbound=0, ubound=float(len(Tcommon)+0), inv=False)
      
    if Tspi == 0.50 or Tspi > 1.1:
        continue
        
    result = []
    # Columns are :  N/1e5,  SpiBulk, Frac, S/N_Bulk, t*, T*, U*
    for j,(aS,U,bestN) in enumerate([\
            (200,7,4), \
            (245,9,0), \
            (290,11,3), \
            (335,12,0), \
            (380,14,2), \
            (425,16,0), \
            (470,17,3)\
            ]):
        try:
            fname = 'dataplots/SPI/{a:03d}/aS{a:03d}_U{U:02d}_Tspi{T:0.2f}.dat'.format(a=aS,U=U,T=Tspi)
            dat = np.loadtxt( fname )
            dat = np.atleast_2d( dat ) 
            # 0 is number
            # 1 is Spi Bulk
            # 3 is Entropy Bulk
            # 4 is t^{*}
            # 5 is T/t^{*} 
            # 6 is U^{*} / t^{*}
            result.append( [ fx(aS), dat[bestN][1], dat[bestN][0], dat[bestN][3] , dat[bestN][5], dat[bestN][6] ] ) 
        except Exception as e:
            print "Did not find aS={:03d}, T={:0.2f}".format(aS,Tspi)
            print e
            pass
    result = np.array(result)
    #print result 
    
    # Choose what goes on the legend title
    Tlabel = np.mean(result[:,4] )
    #~Tlabel = Tspi
    
    if Tspi >= 0.80:
        axS.plot( result[:,0], result[:,3]+ 0.0*i, '.-', ms=3.5, color=color, label='{:0.2f}'.format(Tlabel))
    elif Tspi >= 0.68:
        axS.plot( result[:-2,0], result[:-2,3]+ 0.0*i, '.-', ms=3.5, color=color, label='{:0.2f}'.format(Tlabel))
    elif Tspi >= 0.62:
        axS.plot( result[:-3,0], result[:-3,3]+ 0.0*i, '.-', ms=3.5, color=color, label='{:0.2f}'.format(Tlabel))
    

axS.set_ylabel(r'$\frac{S}{N}$', rotation=0, labelpad=12, fontsize=14 )


for axs in [axS]:
    axs.set_xlim(1., 26.)
    axs.set_xlabel('$U_{0}/t_{0}$', fontsize=14)
    axs.grid(alpha=0.3)

    

    
axS.legend(title='$T/t_{*}$',bbox_to_anchor=(1.15,1.07), loc='upper right', numpoints=1, \
    prop={'size':8}, handlelength=1., handletextpad=0.5)



axS.set_ylim(0.,1.2)

        
figS.savefig('dataplots/SEXP/ExtFig04_S.png', dpi=300)
```
